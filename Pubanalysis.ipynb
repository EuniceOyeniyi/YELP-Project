```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Pubdata.csv')
df.head(2)
```

```python
df['Neighbourhoods'].value_counts().plot(kind = 'bar' )
plt.xlabel('Locations')
plt.ylabel('Counts')
plt.title('Hit location for pubs')
```

```python
df['Neighbourhoods'].value_counts().nlargest(10).plot(kind = 'pie',figsize=(8,8), autopct = '%1.1f%%')
plt.title('Percentage of location of the resturants')
```

```python
df[['Neighbourhoods','Number_of_reviews']].groupby(['Neighbourhoods']).sum().sort_values('Number_of_reviews',ascending=False).plot(kind='bar')
plt.ylabel('Counts')
plt.title(' Mostly visited Neighbourhoods ')
```

```python
new= df[df['Ratings']>=3.5]
new = new.sort_values('Ratings',ascending=False)
```

```python
new[['Neighbourhoods','Ratings']].groupby('Ratings').value_counts().plot(kind='bar', figsize=(12,5))
plt.ylabel('Counts')
plt.title('Rating based on Neighbourhoods')
```

```python
df[['Ratings','Prices']].groupby(['Prices']).value_counts().plot(kind='bar')
plt.ylabel('Counts')
plt.title('Most prefered price range')
```

```python
df[['Prices','Number_of_reviews']].groupby(['Prices']).sum().plot(kind='bar')
plt.ylabel('Counts')
plt.title('Most prefered price range')
```
